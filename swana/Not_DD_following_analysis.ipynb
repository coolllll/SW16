```python
import json
file_data = open('SW16_Follower.json')
json_list =[]
for line in file_data :
    json_list.append(json.loads(line))
dict_follower = {}
for member in json_list :
    tmp = list(member.values())
    for follower in tmp[0] :
        if follower in dict_follower :
            dict_follower[follower] += 1
        else :
            dict_follower[follower] = 1
print(dict_follower)
```

```python
list_one_follow = []
for i,j in dict_follower.items() :
    if j <= 2 :
        list_one_follow.append(i)
print(list_one_follow)
```

```python
counter = 0
dict_member = {}
for i in list_one_follow :
    for member in json_list :
        if i in list(member.values())[0] :
            for name,fl in member.items() :
                counter += 1
                if name in dict_member :
                    dict_member[name] += 1
                else :
                    dict_member[name] = 1
```

```python
import operator
x = sorted(dict_member.items(), key=operator.itemgetter(1) , reverse = True)
print(x)
```

```python
import csv
with open('Most_member_followed_by_not_DD.csv', 'w', newline='') as csvfile:
    fieldnames = ['number of follows', 'count']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    write_dict = {}
    for key,value in x :
        write_dict['number of follows'] = key
        write_dict['count'] = value
        writer.writerow(write_dict)
        counter += value
        print(str(key) + " " + str(value))
```
