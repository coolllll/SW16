```python
import json
file_data = open('SW16_Follower.json')
json_list =[]
for line in file_data :
    json_list.append(json.loads(line))
print(json_list[1])
```

```python
dict_follower = {}
for member in json_list :
    tmp = list(member.values())
    for follower in tmp[0] :
        if follower in dict_follower :
            dict_follower[follower] += 1
        else :
            dict_follower[follower] = 1
print(dict_follower)
```

```python
dict_count = {}
for i in range(1,14) :
    dict_count[i] = 0
for i in dict_follower.values() :
    dict_count[i] += 1
print(dict_count)
```

```python
import matplotlib.pyplot as plt

D = dict_count

plt.bar(range(len(D)), list(D.values()), align='center')
plt.xticks(range(len(D)), list(D.keys()))
# # for python 2.x:
# plt.bar(range(len(D)), D.values(), align='center')  # python 2.x
# plt.xticks(range(len(D)), D.keys())  # in python 2.x

plt.show()
```

```python
import csv
with open('Following_count.csv', 'w', newline='') as csvfile:
    fieldnames = ['number of follows', 'count']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    write_dict = {}
    for key,value in dict_count.items() :
        write_dict['number of follows'] = key
        write_dict['count'] = value
        writer.writerow(write_dict)
        print(str(key) + " " + str(value))
```

```python
import operator
x = sorted(D.items(), key=operator.itemgetter(1) , reverse = True)
print(x)
```

```python
counter = 0
with open('Following_count_sorted.csv', 'w', newline='') as csvfile:
    fieldnames = ['number of follows', 'count']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    for key,value in x :
        write_dict['number of follows'] = key
        write_dict['count'] = value
        writer.writerow(write_dict)
        counter += value
        print(str(key) + " " + str(value))
```
